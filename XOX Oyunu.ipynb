Python diline giriş yaptık şimdi ise sizlerle XOX (diğer isimleriyle SOS, TicTacToe) oyunu yazacağız. 

Oyunumuzu metotlara bölerek sırasıyla metotlarımızı tamamladıktan sonra, main fonksiyonumuzu yazarak oyunu çalışmaya hazır hale getireceğiz.  

Oyunu yazmayı bitirdikten sonra ise en sonda isteyenler için birkaç isteğe bağlı sorumuz olacak. 


### Sırasıyla yazacağımız metotlar

1. oyun_tahtasini_goster
2. oyuncu_harf_secme
3. oynama_sirasi_belirle
4. yeniden_basla
5. harf_yerlestirme
6. kazanani_belirle
7. kopya_oyun_tahtasi
8. hareket_uygun_mu
9. oyuncu_oynasin
10. listeden_random_hareket_sec
11. bilgisayar_oynasin
12. tahta_dolu_mu
13. main metodu

```python
# Tic Tac Toe

import random

def oyun_tahtasini_goster(oyun_tahtasi):
    # This function prints out the oyun_tahtasi that it was passed.

    # "oyun_tahtasi" is a list of 10 strings representing the oyun_tahtasi (ignore index 0)
    print('   |   |')
    print(' ' + oyun_tahtasi[7] + ' | ' + oyun_tahtasi[8] + ' | ' + oyun_tahtasi[9])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + oyun_tahtasi[4] + ' | ' + oyun_tahtasi[5] + ' | ' + oyun_tahtasi[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + oyun_tahtasi[1] + ' | ' + oyun_tahtasi[2] + ' | ' + oyun_tahtasi[3])
    print('   |   |')

def oyuncu_harf_secme():
    # Lets the oyuncu type which harf they want to be.
    # Returns a list with the oyuncu's harf as the first item, and the bilgisayar's harf as the second.
    harf = ''
    while not (harf == 'X' or harf == 'O'):
        print('X mi olmak istersin O mu?')
        harf = input().upper()

    # the first element in the tuple is the oyuncu's harf, the second is the bilgisayar's harf.
    if harf == 'X':
        return ['X', 'O']
    else:
        return ['O', 'X']

def oynama_sirasi_belirle():
    # Randomly choose the oyuncu who goes first.
    if random.randint(0, 1) == 0:
        return 'bilgisayar'
    else:
        return 'oyuncu'

def yeniden_basla():
    # This function returns True if the oyuncu wants to play again, otherwise it returns False.
    print('Tekrar oynamak ister misin? (e veya h)')
    return input().lower().startswith('e')

def harf_yerlestirme(oyun_tahtasi, harf, hareket):
    oyun_tahtasi[hareket] = harf

def kazanani_belirle(o_t, le):
    # Given a oyun_tahtasi and a oyuncu's harf, this function returns True if that oyuncu has won.
    # We use bo instead of oyun_tahtasi and le instead of harf so we don't have to type as much.
    return ((o_t[7] == le and o_t[8] == le and o_t[9] == le) or # across the top
    (o_t[4] == le and o_t[5] == le and o_t[6] == le) or # across the middle
    (o_t[1] == le and o_t[2] == le and o_t[3] == le) or # across the bottom
    (o_t[7] == le and o_t[4] == le and o_t[1] == le) or # down the left side
    (o_t[8] == le and o_t[5] == le and o_t[2] == le) or # down the middle
    (o_t[9] == le and o_t[6] == le and o_t[3] == le) or # down the right side
    (o_t[7] == le and o_t[5] == le and o_t[3] == le) or # diagonal
    (o_t[9] == le and o_t[5] == le and o_t[1] == le)) # diagonal

def kopya_oyun_tahtasi(oyun_tahtasi):
    # Make a duplicate of the oyun_tahtasi list and return it the duplicate.
    kopya_oyun_tahtasi = []

    for i in oyun_tahtasi:
        kopya_oyun_tahtasi.append(i)

    return kopya_oyun_tahtasi

def hareket_uygun_mu(oyun_tahtasi, hareket):
    # Return true if the passed hareket is free on the passed oyun_tahtasi.
    return oyun_tahtasi[hareket] == ' '

def oyuncu_oynasin(oyun_tahtasi):
    # Let the oyuncu type in his hareket.
    hareket = ' '
    while hareket not in '1 2 3 4 5 6 7 8 9'.split() or not hareket_uygun_mu(oyun_tahtasi, int(hareket)):
        print('Siradaki hareket? (1-9)')
        hareket = input()
    return int(hareket)

def listeden_random_hareket_sec(oyun_tahtasi, hareket):
    # Returns a valid hareket from the passed list on the passed oyun_tahtasi.
    # Returns None if there is no valid hareket.
    muhtemel_hareketler = []
    for i in hareket:
        if hareket_uygun_mu(oyun_tahtasi, i):
            muhtemel_hareketler.append(i)

    if len(muhtemel_hareketler) != 0:
        return random.choice(muhtemel_hareketler)
    else:
        return None

def bilgisayar_oynasin(oyun_tahtasi, bilgisayar_harf):
    # Given a oyun_tahtasi and the bilgisayar's harf, determine where to hareket and return that hareket.
    if bilgisayar_harf == 'X':
        oyuncu_harf = 'O'
    else:
        oyuncu_harf = 'X'

    # Here is our algorithm for our Tic Tac Toe AI:
    # First, check if we can win in the next hareket
    for i in range(1, 10):
        kopya = kopya_oyun_tahtasi(oyun_tahtasi)
        if hareket_uygun_mu(kopya, i):
            harf_yerlestirme(kopya, bilgisayar_harf, i)
            if kazanani_belirle(kopya, bilgisayar_harf):
                return i

    # Check if the oyuncu could win on his next hareket, and block them.
    for i in range(1, 10):
        kopya = kopya_oyun_tahtasi(oyun_tahtasi)
        if hareket_uygun_mu(kopya, i):
            harf_yerlestirme(kopya, oyuncu_harf, i)
            if kazanani_belirle(kopya, oyuncu_harf):
                return i

    # Try to take one of the corners, if they are free.
    hareket = listeden_random_hareket_sec(oyun_tahtasi, [1, 3, 7, 9])
    if hareket != None:
        return hareket

    # Try to take the center, if it is free.
    if hareket_uygun_mu(oyun_tahtasi, 5):
        return 5

    # Move on one of the sides.
    return listeden_random_hareket_sec(oyun_tahtasi, [2, 4, 6, 8])

def tahta_dolu_mu(oyun_tahtasi):
    # Return True if every space on the oyun_tahtasi has been taken. Otherwise return False.
    for i in range(1, 10):
        if hareket_uygun_mu(oyun_tahtasi, i):
            return False
    return True


print("UD 18 XOX'e Hos Geldiniz!")

while True:
    # Reset the oyun_tahtasi
    the_tahta = [' '] * 10
    oyuncu_harf, bilgisayar_harf = oyuncu_harf_secme()
    sira = oynama_sirasi_belirle()
    print( sira + ' baslayacak.')
    oyun_devam = True

    while oyun_devam:
        if sira == 'oyuncu':
            # Player's turn.
            oyun_tahtasini_goster(the_tahta)
            hareket = oyuncu_oynasin(the_tahta)
            harf_yerlestirme(the_tahta, oyuncu_harf, hareket)

            if kazanani_belirle(the_tahta, oyuncu_harf):
                oyun_tahtasini_goster(the_tahta)
                print('Tebrikler, oyunu kazandin!')
                oyun_devam = False
            else:
                if tahta_dolu_mu(the_tahta):
                    oyun_tahtasini_goster(the_tahta)
                    print('Oyun berabere, belki bir dahaki sefer kazanirsin!')
                    break
                else:
                    sira = 'bilgisayar'

        else:
            # Computer's turn.
            hareket = bilgisayar_oynasin(the_tahta, bilgisayar_harf)
            harf_yerlestirme(the_tahta, bilgisayar_harf, hareket)

            if kazanani_belirle(the_tahta, bilgisayar_harf):
                oyun_tahtasini_goster(the_tahta)
                print('Bilgisayar kazandi, KAYBETTIN.')
                oyun_devam = False
            else:
                if tahta_dolu_mu(the_tahta):
                    oyun_tahtasini_goster(the_tahta)
                    print('Oyun berabere, belki bir dahaki sefer kazanirsin!')
                    break
                else:
                    sira = 'oyuncu'

    if not yeniden_basla():
        break
```
